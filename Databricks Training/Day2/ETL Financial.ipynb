```python
from pyspark.sql.functions import *
```

```python
df=spark.read.csv("/Volumes/acc_dev/naval/raw/financial_dataset.csv",header=True,inferSchema=True)
df1=df.withColumnsRenamed({"TransactionID":"transactionid","CustomerID":"customerid","AccountType":"accounttype","TransactionType":"transactiontype","AmountUSD":"amountusd","Branch":"branch","TransactionDate":"transactiondate"})
df1=df1.withColumn("ingestion_date",current_timestamp()).withColumn("path",lit("/Volumes/acc_dev/naval/raw/financial_dataset.csv"))
df1.write.mode("overwrite").option("mergeSchema", "true").saveAsTable("acc_dev.munnu.financial_dataset")
```
